```python
import pandas as pd
import numpy as np
import scipy as sp
import scipy.optimize
import numpy.linalg
```

```python
asset_prices = pd.read_csv("Acutal Testing Data.csv", index_col=0)
```

```python
asset_price_predictions_1 = pd.read_csv("Predicted Testing Data Analyst 1.csv", index_col=0)
asset_price_predictions_2 = pd.read_csv("Predicted Testing Data Analyst 2.csv", index_col=0)
asset_price_predictions_3 = pd.read_csv("Predicted Testing Data Analyst 3.csv", index_col=0)
```

```python
def allocate_portfolio(asset_prices, asset_price_predictions_1,asset_price_predictions_2,asset_price_predictions_3):
    
    # This simple strategy equally weights all assets every period
    # (called a 1/n strategy).
    
    n_assets = len(asset_prices)
    n = 9
    weights = np.repeat(1 / n, n)
    return weights
```

```python
linear_weights = allocate_portfolio(asset_prices, asset_price_predictions_1, asset_price_predictions_2, asset_price_predictions_3)

def percentage_change_computed_df(df):
    for x in range(0,9):   
        df['shift'+ str(x)] = df[df.columns[x]].shift(1)
        df['PercentChange'+ str(x)] = (df[df.columns[x]] - df['shift'+ str(x)]).div(df['shift'+ str(x)]) * 100
        df=df.drop(columns=('shift'+str(x)))
    return df

def weighted_percentage_df(df, weights):
    returns = df.loc[:,"PercentChange0":"PercentChange8"]
    return np.dot(returns, weights)

def sharpe(r):
    print(np.mean(r))
    print(np.std(r))
    return np.mean(r)/np.std(r)
```

```python
asset_prices = percentage_change_computed_df(asset_prices)
asset_prices['Total_Returns'] = weighted_percentage_df(asset_prices, linear_weights)
asset_prices
```

```python
# Linear Weights Sharpe
returns = asset_prices.loc[:,"PercentChange0":"PercentChange8"].iloc[1:,:]
sharpe(asset_prices['Total_Returns'])
```

```python
def covmatrix(returns):
    return pd.DataFrame(np.cov(np.transpose(returns)))
```

```python
cov_matrix = covmatrix(returns)
cov_matrix
```

```python
def min_risk_func(x):
    a, b, c, d, e, f, g, h = list(zip(x))
    i = 1-a[0]-b[0]-c[0]-d[0]-e[0]-f[0]-g[0]-h[0]
    weights = np.array([a[0], b[0], c[0], d[0], e[0], f[0], g[0], h[0], i])
    return np.dot(weights, np.dot(cov_matrix, weights))
```

```python
guess = np.repeat(1 / 9, 8)
min_risk_weights = scipy.optimize.minimize(min_risk_func, guess).x
min_risk_weights = np.append(min_risk_weights, (1-sum(min_risk_weights)))
min_risk_weights
```

```python
asset_prices = percentage_change_computed_df(asset_prices)
asset_prices['Total_Returns'] = weighted_percentage_df(asset_prices, min_risk_weights)
asset_prices
```

```python
# Min Risk Weights Sharpe
sharpe(asset_prices['Total_Returns'])
```
